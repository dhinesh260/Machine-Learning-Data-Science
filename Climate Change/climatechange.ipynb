Climate Change Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import numpy as np
sns.set()
```

Importing Data

```python
years_base, means_base = np.loadtxt('5-year-mean-1951-1980.csv', delimiter=',', usecols=(0, 1), unpack=True)
years, means = np.loadtxt('5-year-mean-1882-2014.csv', delimiter=',', usecols=(0, 1), unpack=True)
```

```python
plt.scatter(years_base, means_base)
plt.title('Scatter Plot for Mean Temperature Difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('means', fontsize=12)
plt.show()
```

```python
intercept, slope = np.polyfit(years_base, means_base, 1)
```

```python
print(intercept, slope)
```

```python
def f(x):
    return intercept * x + slope
```

```python
plt.scatter(years_base, means_base)
plt.plot(years_base, f(years_base))
plt.title('Scatter plot of Mean Temp Difference vs Year')
plt.xlabel('Years', fontsize=12)
plt.ylabel('Mean Temp', fontsize=12)
plt.show()
```

```python
print('y = {0} * x + {1}'.format(slope, intercept))
```

```python
model = LinearRegression(fit_intercept=True)
```

```python
model
```

```python
model.fit(years_base[:,np.newaxis], means_base)
```

```python
mean_predicted = model.predict(years_base[:, np.newaxis])
```

```python
mean_predicted
```

```python
plt.scatter(years_base, means_base)
plt.plot(years_base, mean_predicted)
plt.title('Mean Temp Diff vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('Mean Temp diff', fontsize=12)
plt.show()
```

```python
print('y = {0} * x + {1}'.format(model.intercept_, model.coef_[0]))
```
